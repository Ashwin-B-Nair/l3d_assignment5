The model is able to classify object with a fairly high level of accuracy. However, it fails in cases where the objects resemble samples from another class. For example, sample 619 is wrongly classified as a Lamp maybe due to the broad base and long stem which closely resemble the features of the Lamp class. On the other hand, complex structures like the chair in sample 406 is wrongly classified as a Lamp, likely due to the sample having less general features of the chair class. These failure cases shows that the vanilla PointNet network is confused when there are complex shapes or shapes that resemble two different classes closely.

### Robustness Analysis

I analysed two methods to check for robustness, rotation and reducing the number of points in the point cloud. 

`Rotation:` I rotated the input point clouds 45 degrees about the x-axis. 

`Sample points:` I changed the number of sample points from 10,000 to 1000. 


`Classification Model`

When the images are rotated, the model accuracy dropped to 73.13% indicating that PointNet fails to capture the invariance to rotation. 
When the number of point were reduced to 1000, the drop in accuracy was not that significant as it dropped to 97.16%, showing that the model is able to achieve good results even with a moderate amount of points.
